# DUCK CURVE for different seasons for India #

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from timezonefinder import TimezoneFinder
import pvlib
from pvlib.location import Location
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
```

### WINTER (15 January) ###

```python
df1 = pd.read_excel(r"D:\IIT BOMBAY\April 2025\Duck_Curve\Indian_duck_curve_15_Jan.xlsx")
df1.head()
```

```python
# Convert 'Hour' column to proper time format
df1.index = pd.to_datetime(df1["time"], format='%H:%M:%S')

fig, ax = plt.subplots(figsize=(10, 5))

plt.plot(df1.index, df1["Net_Load_2020"], label="2020", color="black", marker="P", linestyle="-")
plt.plot(df1.index, df1["Net_Load_2021"], label="2021", color="yellow", marker="x", linestyle="-")
plt.plot(df1.index, df1["Net_Load_2022"], label="2022", color="blue", marker="o", linestyle="-")
plt.plot(df1.index, df1["Net_Load_2023"], label="2023", color="red", marker="s", linestyle="-")
plt.plot(df1.index, df1["Net_Load_2024"], label="2024", color="green", marker="^", linestyle="-")

date_form = DateFormatter('%H:%M')
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.HourLocator(interval=2))

# Formatting the plot
plt.xlabel("Hour of the Day", fontsize=12)
plt.ylabel("Net Load (MW)", fontsize=12)
plt.title("Duck Curve for India (Winter season)", fontsize=16, weight= "bold")
#plt.xticks(range(0, 24, 1))
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```
